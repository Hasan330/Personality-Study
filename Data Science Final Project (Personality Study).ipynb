# Now we want to calculate personality stats according to the following principles:
https://en.wikipedia.org/wiki/Myers%E2%80%93Briggs_Type_Indicator#/media/File:MyersBriggsTypes.png

We will do that following this logic:

Introvertedness Vs. Extravertedness = 
(4 * Fun with friends) + (-5 * Fear of public speaking) + (3 * Funniness) + (-4 * Loneliness) + (5 * Number of friends) + (5 * New environment) + (4 * Socializing) + (2 * Entertainment Spending) + (4 * Celebrities)

Sensing Vs. Intiution = 
(-4 * Judgement Calls) + (-4 * art) + (-4 * read) + (-3 * Theatre) + (4 * Daily Events) + ( -4 * Writting) + (-3 * Thinking ahead) + (-4 * Interests or hobbies) + (4 * Eating to survive)





```python
allData = dframe.copy()
allData.fillna(3, inplace=True)  #Replacing empty values with the number 3 (The median score)
```

```python
medianScore =3
IntroExtraFactor = (4 * allData.loc[:, "Fun with friends"]) + (-5 * allData.loc[:, "Fear of public speaking"]) + (3 * allData.loc[:, "Funniness"]) + (-4 * allData.loc[:, "Loneliness"]) + (5 * allData.loc[:, "Number of friends"]) + (5 * allData.loc[:, "New environment"])+ (4 * allData.loc[:, "Socializing"]) + (2 * allData.loc[:, "Entertainment spending"]) + (4 * allData.loc[:, "Celebrities"])
threshold        = (4 * medianScore) +                        (-5 * medianScore) +                               (3 * medianScore) +                 (-4 * medianScore) +                  (5 * medianScore) +                         (5 * medianScore)+                       (4 * medianScore) +                   (2 * medianScore) +                              (4 * medianScore)
print("The threshold, above which the person will be considered as an introvert is:", threshold)

maxValue = (4 * 5) +  (-5 * 1) + (3 * 5) + (-4 * 1) +  (5 * 5) +  (5 * 5)+  (4 * 5) +  (2 * 5) + (4 * 5)
difference = maxValue-threshold
print("Max Value=", maxValue, difference)

minValue = (4 * 1) + (-5 * 5) +  (3 * 1) + (-4 * 5) + (5 * 1) + (5 * 1) + (4 * 1) + (2 * 1) + (4 * 1)
#difference = minValue-threshold
print("Min Value=", minValue, difference)
#print(IntroExtraFactor)
```

```python
isExtravert = []
IConfidence = []
EConfidence = []

for i in range (len(IntroExtraFactor)): 
    extravertConfidence = round((IntroExtraFactor[i]-threshold)/difference*100)/2 + 50
    #print("ExtraConfidence=", extravertConfidence, "IntroConfidence=", 100 - extravertConfidence)
    IConfidence.append(100-extravertConfidence)
    EConfidence.append(extravertConfidence)
    if(IntroExtraFactor[i] > threshold):
        #print("Extravert: Person number", i+1, "is", extravertConfidence,"% Extraverted\n")
        isExtravert.append("E")
    else:
        isExtravert.append("I")
        #print("Introvert: Person number", i+1, "is", 100 - extravertConfidence,"% Introverted\n")

allData['isExtravert'] = isExtravert
allData['IConfidence'] = IConfidence
allData['EConfidence'] = EConfidence
        
#allData
```

```python
#Sensing Vs. Intiution = (-4 Judgement Calls) + (-4 art) + (-4 read) + (-3 Theatre) +               (4 Daily Events) + ( -4 Writting) + (-3 Thinking ahead) + (-4 Interests or hobbies) + (4 * Eating to survive)

SensingIntiutionFactor = (-4 * allData.loc[:, "Judgment calls"]) + (-4 * allData.loc[:, "Art exhibitions"]) + (-4 * allData.loc[:, "Reading"]) + (-3 * allData.loc[:, "Theatre"]) + (4 * allData.loc[:, "Daily events"]) + (-4 * allData.loc[:, "Writing"])+ (-3 * allData.loc[:, "Thinking ahead"]) + (-3 * allData.loc[:, "Interests or hobbies"]) + (4 * allData.loc[:, "Eating to survive"])
threshold              = (-4 * medianScore) +                      (-4 * medianScore) +                       (-4 * medianScore) +               (-3 * medianScore) +               (4 * medianScore) +                    (-4 * medianScore)+               (-3 * medianScore) +                      (-3 * medianScore) +                            (4 * medianScore)

maxValue = (-4*1)+ (-4*1)+ (-4*1) + (-3*1)+ (4*5) + (-4*1) + (-3*1) + (-3*1) + (4*5)
minValue = (-4*5)+ (-4*5)+ (-4*5) + (-3*5)+ (4*1) + (-4*5) + (-3*5) + (-3*5) + (4*1)
difference = maxValue-threshold
print(maxValue, minValue, difference)

print("The threshold, above which the person will be considered to be Sensing is:", threshold)
```

```python
isSensing = []
SConfidence = []
NConfidence = []

for i in range (len(SensingIntiutionFactor)): 
    sensingConfidence = round((SensingIntiutionFactor[i]-threshold)/difference*100)/2 + 50
    #print("Sensing Confidence=", sensingConfidence, "Inuition Confidence=", 100 - sensingConfidence)
    NConfidence.append(100-sensingConfidence)
    SConfidence.append(sensingConfidence)
    if(SensingIntiutionFactor[i] > threshold):
        #print("Sensing: Person number", i+1, "is", sensingConfidence,"% Sensing\n")
        isSensing.append("S")
    else:
        isSensing.append("N")
        #print("Intuitive: Person number", i+1, "is", 100 - sensingConfidence,"% Intuition\n")

allData['isSensing'] = isSensing
allData['SConfidence'] = SConfidence
allData['NConfidence'] = NConfidence

#allData
```

```python
ThinkingFeelingFactor = (-5 * allData.loc[:, "Empathy"]) + (3 * allData.loc[:, "Thinking ahead"]) + (-2 * allData.loc[:, "Daily events"]) + (-4 * allData.loc[:, "Keeping promises"]) + (3 * allData.loc[:, "Loss of interest"])+ (-3 * allData.loc[:, "Self-criticism"]) + (-3 * allData.loc[:, "Giving"]) + (-3 * allData.loc[:, "Charity"]) + (-2 * allData.loc[:, "Borrowed stuff"]) + (2 * allData.loc[:, "Assertiveness"])
threshold             = (-5 * medianScore) +               (3 * medianScore) +                      (-2 * medianScore) +                    (-4 * medianScore) +                        (3 * medianScore) +                       (-3 * medianScore)+                       (-3 * medianScore) +              (-3 * medianScore) +               (-2 * medianScore)+                       (2 * medianScore)

maxValue= (-5*1) + (3*5) + (-2*1) + (-4*1) + (3*5) + (-3 *1) + (-3*1) + (-3*1) + (-2*1) + (2*5)
minValue= (-5*5) + (3*1) + (-2*5) + (-4*5) + (3*1) + (-3 *5) + (-3*5) + (-3*5) + (-2*5) + (2*1)
difference = maxValue-threshold
print("The threshold, above which the person will be considered to be a Thinker is:", threshold)
print(maxValue, minValue, difference)
```

```python
isThinking = []
TConfidence = []
FConfidence = []

for i in range (len(ThinkingFeelingFactor)): 
    thinkingConfidence = round((ThinkingFeelingFactor[i]-threshold)/difference*100)/2 + 50
    #print("Thinking Confidence=", thinkingConfidence, "Feeling Confidence=", 100 - thinkingConfidence)
    TConfidence.append(thinkingConfidence)
    FConfidence.append(100 - thinkingConfidence)
    if(ThinkingFeelingFactor[i] > threshold):
        #print("Thinking: Person number", i+1, "is", thinkingConfidence,"% Thinking\n")
        isThinking.append("T")
    else:
        isThinking.append("F")
        #print("Feeling: Person number", i+1, "is", 100 - thinkingConfidence,"% Feeling\n")

allData['isThinking']  = isThinking
allData['TConfidence'] = TConfidence
allData['FConfidence'] = FConfidence

#allData
```

```python
JudgingPerceivingFactor = (4 * allData.loc[:, "Prioritising workload"]) + (4 * allData.loc[:, "Writing notes"]) + (3 * allData.loc[:, "Thinking ahead"])  + (4 * allData.loc[:, "Responding to a serious letter"])+ (-3 * allData.loc[:, "Energy levels"])
threshold               = (4 * medianScore) +                             (4 * medianScore) +                     (3 * medianScore) +                       (4 * medianScore) +                                     (-3 * medianScore)

maxValue                = (4*5) + (4*5) + (3*5) + (4*5) + (-3*1)
minValue                = (4*1) + (4*1) + (3*1) + (4*1) + (-3*5)
difference = maxValue-threshold
print("The threshold, above which the person will be considered to be a Judging is:", threshold)
print(maxValue, minValue, difference)
```

```python
isJudging = []
JConfidence = []
PConfidence = []

for i in range (len(JudgingPerceivingFactor)): 
    judgingConfidence = round((JudgingPerceivingFactor[i]-threshold)/difference*100)/2 + 50
    #print("Judging Confidence=", judgingConfidence, "Perceiving Confidence=", 100 - judgingConfidence)
    PConfidence.append(100-judgingConfidence)
    JConfidence.append(judgingConfidence)
    if(JudgingPerceivingFactor[i] > threshold):
        #print("Judging: Person number", i+1, "is", judgingConfidence,"% Judging\n")
        isJudging.append("J")
    else:
        isJudging.append("P")
        #print("Perceiving: Person number", i+1, "is", 100 - judgingConfidence,"% Perceiving\n")

allData['isJudging'] = isJudging
allData['JConfidence'] = JConfidence
allData['PConfidence'] = PConfidence

#allData
```

# Now let us create a column with personality type in it (Ex: "ESTJ")

```python
#Return Personality as letters (Ex: ESFP)
def PersonalityType(data):
    personality = []
    personalityNumber = []
    for i in range(len(data)):
        personality.append(data['isExtravert'][i]+ data['isSensing'][i] + data['isThinking'][i]+ data['isJudging'][i]) 
        if(data['isExtravert'][i] == 'E'):
            if(data['isSensing'][i] == 'S'):
                if(data['isThinking'][i] == 'T'):
                    if(data['isJudging'][i] == 'J'):
                        personalityNumber.append(1)
                    else:
                        personalityNumber.append(2)
        
        if(data['isExtravert'][i] == 'E'):
            if(data['isSensing'][i] == 'S'):
                if(data['isThinking'][i] == 'F'):
                    if(data['isJudging'][i] == 'J'):
                        personalityNumber.append(3)
                    else:
                        personalityNumber.append(4)
        
        if(data['isExtravert'][i] == 'E'):
            if(data['isSensing'][i] == 'N'):
                if(data['isThinking'][i] == 'T'):
                    if(data['isJudging'][i] == 'J'):
                        personalityNumber.append(5)
                    else:
                        personalityNumber.append(6)
        
        if(data['isExtravert'][i] == 'E'):
            if(data['isSensing'][i] == 'N'):
                if(data['isThinking'][i] == 'F'):
                    if(data['isJudging'][i] == 'J'):
                        personalityNumber.append(7)
                    else:
                        personalityNumber.append(8)
                        
        if(data['isExtravert'][i] == 'I'):
            if(data['isSensing'][i] == 'S'):
                if(data['isThinking'][i] == 'T'):
                    if(data['isJudging'][i] == 'J'):
                        personalityNumber.append(9)
                    else:
                        personalityNumber.append(10)
                        
        if(data['isExtravert'][i] == 'I'):
            if(data['isSensing'][i] == 'S'):
                if(data['isThinking'][i] == 'F'):
                    if(data['isJudging'][i] == 'J'):
                        personalityNumber.append(11)
                    else:
                        personalityNumber.append(12)
        
        if(data['isExtravert'][i] == 'I'):
            if(data['isSensing'][i] == 'N'):
                if(data['isThinking'][i] == 'T'):
                    if(data['isJudging'][i] == 'J'):
                        personalityNumber.append(13)
                    else:
                        personalityNumber.append(14)
        
        
        if(data['isExtravert'][i] == 'I'):
            if(data['isSensing'][i] == 'N'):
                if(data['isThinking'][i] == 'F'):
                    if(data['isJudging'][i] == 'J'):
                        personalityNumber.append(15)
                    else:
                        personalityNumber.append(16)
    
    
    data['Personality Number'] = personalityNumber
    return(personality)
    
        
pers = PersonalityType(allData)
allData['Personality Type'] = pers
allData
```

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
df_left= allData[allData['Left - right handed']=='left handed']
df_right= allData[allData['Left - right handed']=='right handed']
#print(df_right)
x_left, y_left = ecdf(df_left['SConfidence'])
x_right, y_right = ecdf(df_right['SConfidence'])

# Generate plot
_ = plt.plot(x_left, y_left, marker='.', linestyle='none')
_ = plt.plot(x_right, y_right, marker='.', linestyle='none')

# Make the margins nice
plt.margins(0.02)

# Label the axes
_ = plt.legend(('Left-Handed', 'Right-Handed'), loc='lower right')
_ = plt.xlabel('Percentage of Extraverted')
_ = plt.ylabel('ECDF')

# Display the plot
plt.show()
```
